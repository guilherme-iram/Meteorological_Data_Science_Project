```python
import matplotlib.pyplot as plt

cols = [col for col in common_columns if col not in ['TIME', 'DATE']]

non_null_counts = df_final.shape[0] - df_final[cols].isna().sum()

plt.figure(figsize=(10, 6))
bars = plt.bar(non_null_counts.index, non_null_counts.values)
plt.ylim(0, 60000)  
plt.title('BSM - Total de Valores N√£o Nulos por Coluna (Absoluto e Relativo)')
plt.ylabel('Quantidade de Valores N√£o Nulos')
plt.xlabel('Vari√°veis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 500, f'{int(yval)} ({yval/59881 * 100:.2f}%)', ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
colunas_para_analisar = ['TEMPIN', 'TEMPOUT', 'RAIN', 'WINDSPEED', 'BAR']

grupo = df_final.groupby(['YEAR'])

non_null_by_month = grupo[colunas_para_analisar].count().reset_index()

for coluna in colunas_para_analisar:
    x = non_null_by_month['YEAR'].astype(str)
    y = non_null_by_month[coluna]

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, marker='o', label=coluna)
    for xi, yi in zip(x, y):
        plt.text(xi, yi + 200, str(yi), ha='center', va='bottom', fontsize=8)

    plt.title(f'Total de Valores N√£o Nulos por Ano - {coluna}')
    plt.xlabel('Ano')
    plt.ylabel('Contagem de Valores N√£o Nulos')
    plt.xticks(rotation=45)
    plt.ylim(0, df_final[df_final['YEAR'] == 2020].shape[0]) 
    plt.grid(True)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
# C√°lculo da porcentagem de valores faltantes por m√™s para o ano de 2017

anos = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
# Gera√ß√£o dos gr√°ficos com valores absolutos e percentuais
for ano in anos:
    print(ano, '\n')
    df_aux_ano = df_final[df_final['YEAR'] == ano]

    total_por_mes = df_aux_ano.groupby('MONTH')[colunas_para_analisar].size()

    faltantes_por_mes = (
        df_aux_ano.groupby('MONTH')[colunas_para_analisar]
        .apply(lambda x: x.isnull().sum())
    )

    faltantes_pct = faltantes_por_mes.div(total_por_mes, axis=0) * 100

    for coluna in colunas_para_analisar:
        max_pct = faltantes_pct[coluna].max()
        ylim_max = max(110, max_pct + 5)

        plt.figure(figsize=(10, 5))
        plt.bar(faltantes_pct.index, faltantes_pct[coluna], color='skyblue')

        for mes, pct, abs_val in zip(faltantes_pct.index, faltantes_pct[coluna], faltantes_por_mes[coluna]):
            plt.text(mes, pct + 1, f'{pct:.1f}%\n({abs_val})', ha='center', va='bottom', fontsize=8)

        plt.title(f'Dados Faltantes por M√™s em {ano} - {coluna}')
        plt.xlabel('M√™s')
        plt.ylabel('% de Dados Faltantes')
        plt.xticks(range(1, 13))
        plt.ylim(0, ylim_max)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
```

```python
# Atualizar a lista de vari√°veis analisadas
colunas_para_analisar = ['TEMPIN', 'TEMPOUT', 'RAIN', 'WINDSPEED', 'BAR']
anos = sorted(df_final['YEAR'].unique())

# Par√¢metros do layout
n_rows = len(colunas_para_analisar)
n_cols = len(anos) + 1  # √∫ltima coluna ser√° para gr√°fico de linha por vari√°vel
fig_width = n_cols * 5
fig_height = n_rows * 4

fig, axes = plt.subplots(n_rows, n_cols, figsize=(fig_width, fig_height), sharey='row')

if n_rows == 1:
    axes = [axes]
if n_cols == 1:
    axes = [[ax] for ax in axes]

# Preparar dados consolidados por ano
grupo = df_final.groupby(['YEAR'])
non_null_by_year = grupo[colunas_para_analisar].count().reset_index()




# Atualizar anota√ß√µes nos gr√°ficos de linha para exibir % e valor absoluto



# Tamanho total de registros por ano (para calcular %)
total_por_ano = df_final.groupby('YEAR').size()

# Preencher subplots novamente
for row_idx, coluna in enumerate(colunas_para_analisar):
    for col_idx, ano in enumerate(anos):
        ax = axes[row_idx][col_idx]
        df_aux_ano = df_final[df_final['YEAR'] == ano]

        total_por_mes = df_aux_ano.groupby('MONTH')[colunas_para_analisar].size()
        faltantes_por_mes = (
            df_aux_ano.groupby('MONTH')[colunas_para_analisar]
            .apply(lambda x: x.isnull().sum())
        )
        faltantes_pct = faltantes_por_mes.div(total_por_mes, axis=0) * 100

        valores = faltantes_pct[coluna]
        absolutos = faltantes_por_mes[coluna]
        ax.bar(valores.index, valores, color='skyblue')

        for mes, pct, abs_val in zip(valores.index, valores, absolutos):
            ax.text(mes, pct + 1, f'{pct:.1f}%\n({abs_val})', ha='center', va='bottom', fontsize=8)

        ax.set_title(f'{coluna} - {ano}')
        ax.set_xlabel('M√™s')
        if col_idx == 0:
            ax.set_ylabel('% Faltantes')
        ax.set_xticks(range(1, 13))
        ax.set_ylim(0, 110)
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Gr√°fico de linha consolidado com % e valor absoluto
    ax_line = axes[row_idx][-1]
    x = non_null_by_year['YEAR'].astype(str)
    y_abs = non_null_by_year[coluna]
    y_pct = (y_abs / total_por_ano.values) * 100
    ax_line.plot(x, y_pct, marker='o', label=coluna, color='orange')

    for xi, yi_pct, yi_abs in zip(x, y_pct, y_abs):
        ax_line.text(xi, yi_pct + 1, f'{yi_pct:.1f}%\n({yi_abs})', ha='center', va='bottom', fontsize=8)

    ax_line.set_title(f'{coluna} - Total Anual')
    ax_line.set_xlabel('Ano')
    ax_line.set_ylabel('% N√£o Nulos')
    ax_line.set_ylim(0, 110)
    ax_line.grid(True)
    ax_line.legend()

plt.tight_layout()
plt.show()
```

```python
# Preparar figura para s√©ries temporais por vari√°vel (linhas) e ano (colunas) + gr√°fico consolidado

# Definir layout
n_rows = len(colunas_para_analisar)
n_cols = len(anos) + 1  # +1 para o gr√°fico consolidado

fig_width = n_cols * 5
fig_height = n_rows * 4

fig, axes = plt.subplots(n_rows, n_cols, figsize=(fig_width, fig_height), sharey='row')

# Ajustar estrutura de axes
if n_rows == 1:
    axes = [axes]
if n_cols == 1:
    axes = [[ax] for ax in axes]

# Preencher subplots
for row_idx, coluna in enumerate(colunas_para_analisar):
    for col_idx, ano in enumerate(anos):
        ax = axes[row_idx][col_idx]
        df_aux = df_final[df_final['YEAR'] == ano]
        ax.plot(df_aux['DATE'], df_aux[coluna], color='skyblue')
        ax.set_title(f'{coluna} - {ano}')
        ax.set_xlabel('Data')
        if col_idx == 0:
            ax.set_ylabel(coluna)
        ax.grid(True)

    # S√©rie temporal consolidada (√∫ltima coluna)
    ax_cons = axes[row_idx][-1]
    ax_cons.plot(df_final['DATE'], df_final[coluna], color='orange')
    ax_cons.set_title(f'{coluna} - Consolidado')
    ax_cons.set_xlabel('Data')
    ax_cons.grid(True)

plt.tight_layout()
plt.show()
```
