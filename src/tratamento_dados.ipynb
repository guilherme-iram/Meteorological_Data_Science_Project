### Leitura e tratamento das colunas

```python
path_data = "../dados/1_BSM_2017.xlsx"

# Leu todos os DF em forma de dicionário
dfs = pd.read_excel(path_data, sheet_name=None)
```

```python
dfs['Jan17'].head(10)
```

```python
# Ajustando o nome das coluans
for k, df in dfs.items():
    dfs[k] = fix_columns_names(df)
    print(k, dfs[k].shape)
```

```python
# Ajustando o nome das coluans
for k, df in dfs.items():
    dfs[k] = drop_nan_named_columns(df)
    print(k, dfs[k].shape)
```

```python
# Ajustando a coluna Time das coluans
for k, df in dfs.items():
    dfs[k] = clean_time_column(df)
    print(dfs[k].Time[:1])
```

### Padronizando Colunas

```python
# Aplicando a função ao dicionário `dfs`
dfs = standardize_dataframes(dfs)

# Verificando o resultado
for key, df in dfs.items():
    print(f"{key}: shape {df.shape}")
```

### Tratando linhas finais

```python
# Removendo ultimas linhas sem sentido
for k, df in dfs.items():
    print(k)
    print(f"shape {df.shape}")
    dfs[k] = filter_valid_dates(df)
    print(f"shape {dfs[k].shape}")
```

### Tratamento de Datas Faltantes

```python
# Preenchendo os 'buracos' de datas e horas faltantes
for k, df in dfs.items():
    print(k)
    print(f"shape {df.shape}")
    dfs[k] = fill_missing_dates_and_times(df)
    print(f"shape {dfs[k].shape}")
```

```python
# Criar o DataFrame para Junho de 2017 usando um DataFrame de referência
dfs["Jun17"] = generate_empty_month_dataframe(2017, 6, reference_df=dfs["Jan17"])

# Verificar o resultado
print(f"Jun17 criado! Shape: {dfs['Jun17'].shape}")
print(dfs["Jun17"].head())
```

### Unificando os 12 meses em um grande DataFrame

```python
# Nome do arquivo de saída
output_file = "../dados/BSM_2017_clean.xlsx"

# Executar a função e salvar o arquivo
final_df = merge_and_save_dataframes(dfs, output_file)

print(f"Arquivo salvo como {output_file}")
print(f"Shape final do DataFrame: {final_df.shape}")
```
