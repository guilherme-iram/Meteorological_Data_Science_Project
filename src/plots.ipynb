```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
```

```python
path_final_data = "..\dados\\BSM_2017_clean.xlsx"
final_df = pd.read_excel(path_final_data)
```

```python
numeric_columns = ['Bar', 'Cool D-D', 'Heat D-D', 'Hi Temp', 'Hi.1 Speed', 'In .1 Heat',
          'In .1 Heat', 'In .2 EMC', 'In Air Density', 'In Hum',
       'In Temp', 'Low Temp', 'Rain', 'Rain Rate', 'Soil 2 Moist,',
       'Soil Temp 1', 'Temp Out', 'Wind Speed', 'Wind.1 Run']

fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(12, 3 * len(numeric_columns)))

if len(numeric_columns) == 1:
    axes = [axes]

for ax, col in zip(axes, numeric_columns):
    ax.plot(final_df["Date"], final_df[col], label=col, linewidth=1)
    ax.set_title(f"Flutuação - {col} - 2017")
    ax.set_ylabel(col)
    ax.xaxis.set_major_locator(mdates.MonthLocator())  # Mostrar apenas os meses no eixo x
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))  # Formato "Mês"
    ax.tick_params(axis='x', rotation=45)  # Rotacionar rótulos para melhor legibilidade
    ax.legend()

plt.tight_layout()
plt.show()
```
